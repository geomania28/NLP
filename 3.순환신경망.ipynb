# 순환 신경망(Recurrent Neural Network, RNN)



## 1. 순환 신경망의 개념
- 순환 신경망(Recurrent Neural Network, RNN)은 시퀀스 데이터를 처리할 수 있도록 설계된 인공신경망의 한 종류입니다.<br>
 일반적인 신경망과 달리, RNN은 각 단계에서 이전 단계의 정보를 기억하고 이를 다음 단계로 전달하여 시계열이나 자연어 처리와 같이 **순차적 맥락**이 중요한 문제에서 뛰어난 성능을 발휘합니다.<br> 이를 통해 문장, 소리, 동영상 등의 연속된 데이터의 패턴을 학습할 수 있습니다. RNN은 순환 신경망 계열의 알고리즘 중에서 가장 기본적인 형태이며, 좀 더 발전된 형태로 LSTM, GRU 등의 신경망 알고리즘이 있습니다.

 <img src="https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FIe6mK%2FbtrrYR2JYzE%2FYBHMBjxLskjI7zETkmgrX0%2Fimg.png" width="600">

- 순환 신경망(RNN)에서는 데이터의 입력과 출력 형태에 따라 네트워크를 여러 가지 유형으로 나눌 수 있습니다. 이들 유형은 주로 시퀀스 데이터의 특성에 맞게 사용되며, 각 유형의 예시는 다음과 같습니다.
  1. One-to-Many : 하나의 입력이 주어졌을 때 여러 개의 출력을 생성하는 구조입니다.

  - 예시: 이미지 설명 생성 모델(image captioning)이 대표적인 예입니다. 이미지 한 장을 입력으로 받아 설명을 생성하기 위해 여러 단어로 이루어진 시퀀스 출력이 필요합니다.
  
  - RNN에서의 사용: RNN은 입력 데이터를 통해 상태를 생성한 후, 이를 바탕으로 시퀀스 형태의 출력을 하나씩 생성하는 방식으로 구현됩니다.
  
  2. Many-to-One : 여러 개의 입력 시퀀스를 받아 단일 출력을 생성하는 구조입니다.
  - 예시: 감정 분석(sentiment analysis)이 이에 해당합니다. 예를 들어, 리뷰나 문장의 여러 단어가 입력으로 주어지고, 이 문장이 긍정적인지 부정적인지를 나타내는 하나의 클래스 레이블이 출력됩니다.
  - RNN에서의 사용: RNN이 전체 입력 시퀀스를 처리한 후, 마지막 시간 단계에서의 상태를 바탕으로 단일 출력을 생성하게 됩니다.
  
  3. Many-to-Many : 입력 시퀀스에 대해 출력 시퀀스를 생성하는 구조입니다. RNN의 가장 일반적인 형태로, 시퀀스의 길이에 맞춰 입력과 출력을 각각 여러 개 받을 수 있습니다.
  - 예시: 기계 번역(machine translation)이나 자막 생성 등이 이에 해당합니다. 예를 들어, 영어 문장이 입력으로 주어지면 해당되는 다른 언어의 번역된 문장이 출력으로 생성됩니다.
  - RNN에서의 사용: 전체 입력 시퀀스를 순차적으로 처리하면서, 각 단계에서 대응하는 출력을 생성하거나, 입력이 끝난 후에 출력 시퀀스를 생성하는 방식으로 구현됩니다.

## 2. 순환 신경망의 등장 배경
 - RNN이 등장한 배경에는 시간에 따른 데이터의 변화나 순서에 의존하는 문제를 해결하기 위한 필요성이 자리하고 있습니다. 기존의 전방향 신경망(FNN)은 각 입력을 독립적으로 처리해 시간적 의존성을 다룰 수 없었습니다. 반면, 언어나 음성 데이터와 같은 연속된 정보를 다루기 위해서는 이전 입력이 현재 입력에 영향을 미치는 구조가 필요했으며, 이를 위해 RNN이 고안되었습니다.

## 3. 순환 신경망의 특징
 - RNN의 주요 특징은 내부에 순환 구조가 있어 이전 단계의 출력을 현재 단계의 입력으로 사용하는 "순환" 특성입니다. 이로 인해 정보가 순차적으로 전달되며, 시점 간의 연관성을 파악하는 데 유리합니다.

 - RNN의 또 다른 특징은 파라미터 공유입니다. 시간의 흐름에 따른 파라미터 수가 증가하지 않으므로, 데이터가 길어져도 비교적 효율적으로 모델을 훈련할 수 있습니다. 그러나 긴 시퀀스에서는 기울기 소실 문제(Vanishing Gradient Problem)가 발생하기 쉬워, 정보가 멀리 전달되는 데 한계가 있다는 단점도 존재합니다.

## 4. 다른 신경망과의 차이점
 - RNN은 시퀀스 데이터의 순서를 고려하는 구조로, 기존의 전방향 신경망(FNN)과 차별화됩니다. FNN은 각 입력을 독립적으로 처리하지만, RNN은 이전 단계의 출력을 다음 단계의 입력으로 연결하여 시계열적 패턴을 학습합니다. CNN(합성곱 신경망)은 이미지와 같은 2D 데이터를 처리하는 데 최적화되어 있어 공간적 패턴을 학습하는 데 강점을 보이지만, 시계열과 같이 시간적 순서가 중요한 데이터 처리에는 적합하지 않습니다.

- 추가로, 긴 시퀀스를 처리할 때 LSTM(Long Short-Term Memory)나 GRU(Gated Recurrent Unit)와 같은 구조가 RNN의 단점을 보완하여, 보다 긴 시간의 의존성을 학습할 수 있도록 도와줍니다.

## RNN 코드 예제

```python
# 필요한 라이브러리 import
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# 하이퍼파라미터 설정
max_features = 10000  # 사용할 단어의 최대 개수 (가장 많이 등장하는 단어들)
max_len = 500         # 시퀀스 길이를 최대 500단어로 제한
embedding_dim = 32    # 임베딩 차원

# IMDB 데이터셋 로드 (train/test 분리된 상태로 제공됨)
# num_words로 가장 자주 등장하는 단어 수를 제한
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# 패딩을 추가하여 시퀀스 길이를 동일하게 맞춤 (길이 제한을 초과하는 경우 잘라내고, 짧은 경우 0으로 패딩)
x_train = pad_sequences(x_train, maxlen=max_len)
x_test = pad_sequences(x_test, maxlen=max_len)
```

```python
def get_rnn_model():
    # 모델 생성 (순차 모델)
    model = Sequential()
    # 임베딩 층: 단어 인덱스를 임베딩 벡터로 변환, input_length를 설정하여 시퀀스 길이를 맞춤
    model.add(Embedding(input_dim=max_features, output_dim=embedding_dim, input_length=max_len))
    # SimpleRNN 층: RNN 셀을 추가하여 시퀀스를 순차적으로 처리
    model.add(SimpleRNN(units=64, return_sequences=True))  # units는 출력 차원
    # 드롭아웃으로 과적합 방지
    model.add(Dropout(0.3))
    # SimpleRNN 층: RNN 셀을 추가하여 시퀀스를 순차적으로 처리
    model.add(SimpleRNN(units=32))
    # 드롭아웃으로 과적합 방지
    model.add(Dropout(0.3))
    # 출력 층: 감정 분석을 위해 이진 분류 (긍정/부정) 설정
    model.add(Dense(1, activation='sigmoid'))
    # 모델 컴파일: 이진 분류 문제이므로 binary_crossentropy 사용
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
    # 모델 요약 출력

rnn_model = get_rnn_model()
rnn_model.summary()

# 모델 학습
# validation_split로 훈련 데이터의 일부를 검증 데이터로 사용, epochs와 batch_size는 모델 성능에 맞게 조정 가능
rnn_history = rnn_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_split=0.2)

# 테스트 데이터 평가
rnn_test_loss, rnn_test_acc = rnn_model.evaluate(x_test, y_test)
print(f"테스트 정확도: {rnn_test_acc}") # 0.8154399991035461
```

```python
rnn_acc = rnn_history.history['accuracy']          # 훈련 데이터의 정확도
rnn_val_acc = rnn_history.history['val_accuracy']   # 검증 데이터의 정확도
rnn_loss = rnn_history.history['loss']              # 훈련 데이터의 손실 값
rnn_val_loss = rnn_history.history['val_loss']      # 검증 데이터의 손실 값

# 에포크 수 계산
epochs = range(1, len(rnn_acc) + 1)

# 정확도 시각화
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)  # 1행 2열 중 첫 번째 그래프
plt.plot(epochs, rnn_acc, label='Training rnn_Accuracy')
plt.plot(epochs, rnn_val_acc, label='Validation rnn_Accuracy')
plt.title('Training and Validation rnn_Accuracy')
plt.xlabel('Epochs')
plt.ylabel('rnn_Accuracy')
plt.legend()

# 손실 시각화
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째 그래프
plt.plot(epochs, rnn_loss, label='Training Loss')
plt.plot(epochs, rnn_val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

# 장단기 메모리 (Long Short-Term Memory, LSTM)

## 1.  LSTM의 등장 배경
- LSTM(Long Short-Term Memory)은 전통적인 RNN(Recurrent Neural Network) 구조에서 발생하는 장기 의존성 문제와 기울기 소실 문제를 해결하기 위해 개발된 특별한 유형의 RNN입니다.
기본적인 RNN은 짧은 시퀀스에서는 잘 동작하지만, 시퀀스가 길어질수록 이전 정보가 제대로 전달되지 않아 장기 의존성(Long-Term Dependency) 정보를 학습하기 어려운 단점이 있습니다.
LSTM은 이러한 문제를 해결하는 구조로 설계되었습니다.

## 2. LSTM의 구조
 - LSTM의 핵심은 **셀 상태(cell state)**와 **게이트(gates)**입니다.
 셀 상태는 시퀀스의 모든 타임스텝을 거쳐 흐르는 정보의 주요 통로이며, 불필요한 정보를 게이트가 걸러내고 중요한 정보만 남기는 역할을 합니다.
  이 때문에 필요한 정보는 시퀀스가 길어져도 잘 전달됩니다.

  LSTM은 다음과 같은 세 가지 게이트를 가지고 있습니다:

  - 입력 게이트(Input Gate): 새로운 정보가 셀 상태에 얼마나 추가될지를 결정합니다.
  - 망각 게이트(Forget Gate): 이전 셀 상태에서 어떤 정보를 버릴지를 결정합니다.
  - 출력 게이트(Output Gate): 현재 타임스텝에서 출력할 정보를 결정합니다.

## 3. LSTM의 동작 과정

LSTM은 셀 상태(cell state)와 세 가지 게이트(망각 게이트, 입력 게이트, 출력 게이트)를 통해 정보를 처리하고 장기 의존성을 학습합니다.

1. 망각 게이트 (Forget Gate)
망각 게이트는 이전 타임스텝의 셀 상태에 어떤 정보를 남기고 어떤 정보를 버릴지 결정합니다. 현재 입력 \( x_t \)와 이전 은닉 상태 \( h_{t-1} \)을 받아들여 시그모이드 함수로 변환해 값을 0과 1 사이로 조정하며, 0은 "완전히 버리기", 1은 "완전히 유지하기"를 의미합니다.

$$
f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)
$$

여기서 $f_t$는 각 정보의 유지 여부를 나타내는 벡터입니다.

2. 입력 게이트 (Input Gate)
입력 게이트는 현재 타임스텝에서 들어오는 정보를 셀 상태에 얼마나 추가할지를 결정합니다. 이를 위해 시그모이드 함수와 하이퍼볼릭 탄젠트 함수를 사용합니다.

 - 1. 시그모이드 함수로 중요한 정보 위치를 결정합니다.
   $$
   i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)
   $$
   
 - 2. 하이퍼볼릭 탄젠트 함수를 통해 새로운 후보 셀 상태를 만듭니다.
   $$
   \tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C)
   $$

 - 입력 게이트의 최종 결과는 \( i_t \)와 \( \tilde{C}_t \)의 곱으로, 새로운 정보가 얼마나 중요한지를 나타내는 벡터입니다.



 3. 셀 상태 업데이트
 - 이제 셀 상태를 업데이트합니다. 이전 셀 상태에 망각 게이트의 출력 \( f_t \)를 곱해 일부 정보는 잊고, 입력 게이트의 결과를 더해 새로운 정보를 추가합니다.

$$
C_t = f_t * C_{t-1} + i_t * \tilde{C}_t
$$



 4. 출력 게이트 (Output Gate)
출력 게이트는 다음 타임스텝에 전달할 은닉 상태 \( h_t \)를 결정합니다.

  1. 시그모이드 함수를 통해 출력할 정보의 중요도를 결정합니다.
   $$
   o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)
   $$

  2. 현재 셀 상태 \( C_t \)에 하이퍼볼릭 탄젠트를 적용해 \( -1 \)과 \( 1 \) 사이로 제한하고, 여기에 \( o_t \)를 곱해 최종 은닉 상태를 만듭니다.
   $$
   h_t = o_t * \tanh(C_t)
   $$

이 은닉 상태 \( h_t \)는 출력으로 반환되며, 다음 타임스텝으로 전달됩니다.

## 4.LSTM의 장점
- 장기 의존성: 셀 상태가 정보의 주요 통로가 되어, 중요한 정보를 여러 타임스텝에 걸쳐 유지할 수 있어 장기 의존성 문제를 해결합니다.
기울기 소실 문제 해결: RNN에서 자주 발생하는 기울기 소실 문제를 방지하여, 더 깊고 긴 시퀀스에서도 안정적으로 학습이 가능합니다.

## 5.LSTM의 단점
- 복잡성: 일반 RNN에 비해 구조가 복잡하여 계산량이 많고, 훈련 시간과 메모리 요구 사항이 높습니다.
과적합 가능성: 파라미터 수가 많아 과적합의 위험이 있으며, 드롭아웃과 같은 과적합 방지 기법을 함께 적용하는 것이 중요합니다.

## 6.LSTM의 활용
- LSTM은 텍스트 생성, 번역, 시계열 예측, 감정 분석, 음성 인식 등 시퀀스 데이터를 다루는 다양한 분야에서 사용됩니다.

```python
def get_lstm_model():
    # 모델 생성 (순차 모델)
    model = Sequential()
    # 임베딩 층: 단어 인덱스를 임베딩 벡터로 변환, input_length를 설정하여 시퀀스 길이를 맞춤
    model.add(Embedding(input_dim=max_features, output_dim=embedding_dim, input_length=max_len))
    # RNN 층: RNN 셀을 추가하여 시퀀스를 순차적으로 처리
    model.add(LSTM(units=64, return_sequences=True))  # units는 출력 차원
    # 드롭아웃으로 과적합 방지
    model.add(Dropout(0.3))
    # LSTM 층: LSTM 셀을 추가하여 시퀀스를 순차적으로 처리
    model.add(LSTM(units=32))
    # 드롭아웃으로 과적합 방지
    model.add(Dropout(0.3))
    # 출력 층: 감정 분석을 위해 이진 분류 (긍정/부정) 설정
    model.add(Dense(1, activation='sigmoid'))
    # 모델 컴파일: 이진 분류 문제이므로 binary_crossentropy 사용
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
    # 모델 요약 출력

lstm_model = get_lstm_model()
lstm_model.summary()

# 모델 학습
# validation_split로 훈련 데이터의 일부를 검증 데이터로 사용, epochs와 batch_size는 모델 성능에 맞게 조정 가능
lstm_history = lstm_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_split=0.2)

# 테스트 데이터 평가
lstm_test_loss, lstm_test_acc = lstm_model.evaluate(x_test, y_test)
print(f"테스트 정확도: {lstm_test_acc}")
```

# GRU(Gated Recurrent Unit)

GRU는 LSTM(Long Short-Term Memory) 모델을 단순화한 형태의 RNN(Recurrent Neural Network)입니다.

LSTM의 복잡한 구조를 줄이고 연산 속도를 높이기 위해 2014년에 처음 제안되었습니다.

## 등장 배경

전통적인 RNN은 시퀀스 데이터에서 유용한 정보를 학습할 수 있지만, **기울기 소실 문제**로 인해 긴 시퀀스를 다루기 어려운 단점이 있었습니다. LSTM은 이러한 문제를 해결했지만, 여러 게이트와 상태를 사용하여 구조가 복잡하고 연산 비용이 높았습니다. 이를 개선하기 위해 GRU가 개발되었으며, LSTM보다 간단한 구조로 비슷한 성능을 발휘할 수 있습니다.

## GRU의 특징

- **단순화된 구조**: GRU는 LSTM의 셀 상태(cell state)를 없애고 은닉 상태(hidden state) 하나만 사용합니다. 또한, 게이트 수를 줄여 LSTM의 3개 게이트 대신 **2개의 게이트(리셋 게이트와 업데이트 게이트)**만 사용합니다.
- **빠른 연산 속도**: 구조가 단순하므로 LSTM보다 학습 속도가 빠르며, 더 적은 메모리를 사용합니다.
- **효율적인 성능**: LSTM에 비해 성능 저하가 거의 없이 장기 의존성을 학습할 수 있습니다.

## GRU의 장단점

**장점**
- **빠른 학습 속도**: 구조가 단순하므로, LSTM에 비해 학습 속도가 빠르고 메모리 사용량이 적습니다.
- **장기 의존성 처리**: 긴 시퀀스 데이터에서도 정보를 유지하고 학습할 수 있습니다.

**단점**
- **복잡한 데이터에는 부적합**: GRU는 LSTM에 비해 단순화된 구조이므로, 데이터가 매우 복잡하거나 긴 시퀀스에서 LSTM만큼의 성능을 내기 어려울 수 있습니다.
- **유연성 부족**: GRU는 상태가 적기 때문에, 다양한 상태 정보가 필요한 작업에는 LSTM보다 제한적일 수 있습니다.

## GRU의 작동 원리

GRU는 **리셋 게이트(reset gate)**와 **업데이트 게이트(update gate)**의 두 가지 게이트로 구성되어 있으며, 이를 통해 정보를 선택적으로 기억하거나 잊어버릴 수 있습니다.

### 1. 업데이트 게이트 (Update Gate)
업데이트 게이트는 이전 은닉 상태를 얼마나 유지할지를 결정하는 게이트입니다. 현재 입력 \( x_t \)와 이전 은닉 상태 \( h_{t-1} \)를 받아들여 시그모이드 함수로 변환합니다. 값은 0과 1 사이에 있으며, 1은 "모두 유지", 0은 "모두 잊기"를 의미합니다.

$$
z_t = \sigma(W_z \cdot [h_{t-1}, x_t] + b_z)
$$

여기서 \( z_t \)는 각 정보의 유지 여부를 나타내는 벡터입니다.



### 2. 리셋 게이트 (Reset Gate)
리셋 게이트는 이전 은닉 상태를 얼마나 잊을지 결정합니다. 현재 입력 \( x_t \)와 이전 은닉 상태 \( h_{t-1} \)를 받아들여 시그모이드 함수로 변환하며, 0에 가까울수록 과거 정보를 덜 사용합니다.

$$
r_t = \sigma(W_r \cdot [h_{t-1}, x_t] + b_r)
$$

여기서 \( r_t \)는 이전 정보의 중요성을 결정하는 벡터입니다.



### 3. 새로운 후보 은닉 상태 (Candidate Hidden State)
리셋 게이트 \( r_t \)의 값을 이용해 이전 은닉 상태를 조정한 후, 하이퍼볼릭 탄젠트(tanh) 함수를 사용하여 새로운 후보 은닉 상태 \( \tilde{h}_t \)를 계산합니다.

$$
\tilde{h}_t = \tanh(W_h \cdot [r_t * h_{t-1}, x_t] + b_h)
$$



### 4. 최종 은닉 상태 (Hidden State)
최종 은닉 상태 \( h_t \)는 이전 은닉 상태 \( h_{t-1} \)와 새로운 후보 은닉 상태 \( \tilde{h}_t \)를 업데이트 게이트 \( z_t \)를 통해 가중합하여 결정됩니다. 업데이트 게이트는 이전 상태와 새로운 상태 중 어떤 정보를 더 많이 반영할지를 결정합니다.

$$
h_t = (1 - z_t) * h_{t-1} + z_t * \tilde{h}_t
$$



이 최종 은닉 상태 \( h_t \)가 출력으로 반환되며, 다음 타임스텝으로 전달됩니다.



GRU는 단순한 구조에도 불구하고 강력한 성능을 발휘하여, LSTM과 더불어 시퀀스 데이터를 처리하는 대표적인 RNN 모델로 널리 사용됩니다.


```python
def get_gru_model():
    # 모델 생성 (순차 모델)
    model = Sequential()
    # 임베딩 층: 단어 인덱스를 임베딩 벡터로 변환, input_length를 설정하여 시퀀스 길이를 맞춤
    model.add(Embedding(input_dim=max_features, output_dim=embedding_dim, input_length=max_len))
    # RNN 층: RNN 셀을 추가하여 시퀀스를 순차적으로 처리
    model.add(GRU(units=64, return_sequences=True))  # units는 출력 차원
    # 드롭아웃으로 과적합 방지
    model.add(Dropout(0.3))
    # GRU 층: GRU 셀을 추가하여 시퀀스를 순차적으로 처리
    model.add(GRU(units=32))
    # 드롭아웃으로 과적합 방지
    model.add(Dropout(0.3))
    # 출력 층: 감정 분석을 위해 이진 분류 (긍정/부정) 설정
    model.add(Dense(1, activation='sigmoid'))
    # 모델 컴파일: 이진 분류 문제이므로 binary_crossentropy 사용
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
    # 모델 요약 출력

gru_model = get_gru_model()
gru_model.summary()

# 모델 학습
# validation_split로 훈련 데이터의 일부를 검증 데이터로 사용, epochs와 batch_size는 모델 성능에 맞게 조정 가능
gru_history = gru_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_split=0.2)

# 테스트 데이터 평가
gru_test_loss, gru_test_acc = gru_model.evaluate(x_test, y_test)
print(f"테스트 정확도: {gru_test_acc}")
```
